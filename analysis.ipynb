##### Cell 2/8: Defining the functions

```python
## defining functions

# This function assesses whether a certain value indexed from the data is actually not a 
# Numpy NaN or a Pandas NaN, or a general None, some nones of a different type are still 
# overlooked and other NaN's are processed later
def not_empty(val):
    try: # If parseable:
        if val == np.isnan(val) or pd.isna(val) or val == None: # check if it is a NaN
            return 'Missing' # then return 'Missing'
        else: 
            return val # otherwise return the original value
    except TypeError: # If it a typeError (non-parseable)
        return val # Skip for later processing by returning the original value

# This function is used to index a 'two layered' data index e.g. county['health']['mental health']
double_idx = lambda data, idx, first : data[idx[first]][idx[first+1]]

# This function is used to index a 'two layered' data index e.g. county['health']['mental health']['2020]
triple_idx = lambda data, idx, first : data[idx[first]][idx[first+1]][idx[first+2]]

# This function is specifically for handling indexings with 7 separate elements
def length_7(data, idx):
    if idx[0] == 'bls': # if it's a bls index:
        val = triple_idx(data, idx, 0) / triple_idx(data, idx, 0) # process as follows: triple and triple
        return not_empty(val)
    else:
        if idx[2] == '-': # with a minus do a subtraction
            val = double_idx(data, idx, 0) - double_idx(data, idx, 3) / idx[6] # from a double indexing and another double indexing, with a subtraction
            return not_empty(val) # check that value is not empty
        else:
            val = double_idx(data, idx, 0) / double_idx(data, idx, 3) / idx[6] # from a double indexing and another double indexing, with a divison
            return not_empty(val) # check that value is not empty

class indexed: # make a new class for processing the indexing of the data
    def __init__(self, idx, data): # arguments for initialising are the indexing and the data
        self.l, self.data, self.idx = len(idx), data, idx # define l for length (of indexing), data for data, and idx for the index

        if self.l == 1: # for an index of length 1:
            self.val = self.data[ self.idx[0] ] # take the value
        elif self.l == 2: # for an index of length 2:
            self.val = double_idx(self.data, self.idx, 0) # take a double
        elif self.l == 3: # for an index of length 3:
            self.val = self.data[ self.idx[0] ] / self.data[ self.idx[2] ] # take a single value divided by a single
        elif self.l == 5: # for an index of length 5:
            self.val = double_idx(self.data, self.idx, 0) + double_idx(self.data, self.idx, 3) # take a double plus a double
        elif self.l == 7: # for an index of length 7:
            self.val = length_7(self.data, self.idx) # use the 7 processing function
        else: # for an index of length 16:
            self.val = double_idx(self.data, self.idx, 0) + double_idx(self.data, self.idx, 3) + double_idx(self.data, self.idx, 6) + double_idx(self.data, self.idx, 9) / double_idx(self.data, self.idx, 12) / 10000 # use the following processing order (specific)

def process_indexing(data, idx): # process the indexing
    idx = not_empty( indexed(idx, data ) ) # find the value using the indexed class and check that it is not empty
    return idx.val # return its value property

def get_values( col_idx ): # get the values for the different columns (features)
    col_vals = {} # empty dictinary
    for col in col_idx: # for every column:
        col_vals[col] = [] # make that column name a key
        for state in states_counties: # for that key for every state
            for county in states_counties[state]: # for that state for every county
                try: # if it can be indexed succsfully add this value to the list for that col_vals dictionary
                    value = process_indexing( data[ state ][ county ] , col_idx[ col ] )
                except (TypeError, KeyError, ZeroDivisionError):
                    value = 0 # For these empty values add a zero.
                col_vals[ col ].append( value ) # add value to this col list
    return col_vals # return the columns with names : values

def get_fips_states_counties_index():
    # init variables
    fips = []
    states_i = []
    counties_i = []
    for state in states_counties: # for each state
        for county in states_counties[state]: # for each county
            fips.append( data[state][county]['fips'] ) # add fip
            states_i.append( state ) # add state
            counties_i.append( county ) # add county
    return [ fips, states_i, counties_i ] # return those values

def impute_values(data, cols):
    for state in data.state.unique():
        for col in cols:
            state_mean = np.mean(data.loc[(data[col] != 'Missing') & (data.state == state), col]) # get the mean of the state
            idx = data[(data[col] == "Missing") & (data.state == state)].index # get missing values
            try:
                # print(state_mean)
                data.loc[(data[col] == 'Missing') & (data.state == state), col] = state_mean # make missing values the state mean
            except ValueError:
                print(state_mean, 'VALUEERROR') # for improving function finding value Errors
                continue

def double_mean_centering(X_dc):
    global_mean = np.mean(np.mean(X_dc)) # calculate global mean
    cols = range(len(X_dc[0,:])) # get all the cols
    rows = range(len(X_dc[:,0])) # get all the rows
    # calculate means
    col_means = [ np.mean(X_dc[:,col]) for col in cols ] # find the means for the cols
    row_means = [ np.mean(X_dc[row,:]) for row in rows ] # find the means for the rows
    # subtract means row-wise + column-wise + global-mean
    X_dc = np.transpose([ X_dc[:,col] - col_means[col] for col in cols ]) # subtracting
    X_dc = [ X_dc[row,:] - row_means[row] for row in rows ] # substracting
    X_dc = X_dc + global_mean # adding
    return X_dc
```

##### Cell 3/8: Loading in the data

```python
### Loading in the data
data = json.load( open('counties.json') )
states = data.keys()
states_counties = {state : list(data[state]) for state in states}

# Define column names for input and output variables
input_colnames = [  'land_area',  
                    'longitude', 
                    'latitude', 
                    'noaa_snow', 
                    'noaa_temp', 
                    'noaa_altitude', 
                    'white', 
                    'black',
                    'asian', 
                    'hispanic',
                    'male_female_ratio', 
                    'pop_trend', 
                    'suicides', 
                    'avg_income', 
                    'uneducated', 
                    'living_cost',
                    'health', 
                    'physically unhealthy (days)', 
                    'mentally unhealthy (days)', 
                    'low birthweight',
                    'excess drinking' ]

output_colnames = [ 'homicides', 
                    'unemployment', 
                    'pol_shootings', 
                    'pol_deaths', 
                    'poverty_rate', 
                    'child poverty', 
                    'crime_rate' ]

# Define indexing for the input and output variables
input_indexing = [  ['land_area'], 
                    ['longitude'], 
                    ['latitude'], 
                    ['noaa', 'snow'],
                    ['noaa', 'temp'],
                    ['noaa', 'altitude'],
                    ['race_demographics', 'non_hispanic_white_alone_male', '+', 'race_demographics', 'non_hispanic_white_alone_female'],
                    ['race_demographics', 'black_alone_male', '+', 'race_demographics', 'black_alone_female'],
                    ['race_demographics', 'asian_alone_male', '+', 'race_demographics', 'asian_alone_female'],
                    ['race_demographics', 'hispanic_male', '+', 'race_demographics', 'hispanic_female'],
                    ['male', '/', 'female'],
                    ['population', '2011', '-', 'population','2019', '/', 8],
                    ['deaths', 'suicides', '/', 'population', '2019', '/', 10000],
                    ['avg_income'],
                    ['edu', 'less-than-high-school'],
                    ['cost-of-living', 'living_wage'],
                    ['health', '% Fair or Poor Health'],
                    ['health', 'Average Number of Physically Unhealthy Days'],
                    ['health', 'Average Number of Mentally Unhealthy Days'],
                    ['health', '% Low birthweight'],
                    ['health', '% Excessive Drinking'] ]

output_indexing = [ ['deaths', 'homicides', '/', 'population', '2019', '/', 10000 ],
                    ['bls', '2020', 'unemployed', '/', 'bls', '2020', 'labor_force'],
                    ['fatal_police_shootings','total-2017', '+', 'fatal_police_shootings', 'total-2018', '+', 'fatal_police_shootings', 'total-2019', '+', 'fatal_police_shootings', 'total-2020', '/', 'population', '2019', '/', 10000 ],
                    ['police_deaths', '/', 'num_police'],
                    ['poverty-rate'],
                    ['health', '% Children in Poverty'],
                    ['health', 'Violent Crime Rate'] ]

# from a dictionary of { colnames : indexing } get the values for the different indexings 
# for each column and for each county
input_col_vals = get_values( dict(zip(input_colnames, input_indexing)) )

output_col_vals = get_values( dict(zip(output_colnames, output_indexing)) )

# Get the fips, states and counties for the index of the dataframe
[fips, states_i, counties_i] = get_fips_states_counties_index()

# Make Series of the states and counties
states_S = pd.Series(states_i, index = fips)

counties_S = pd.Series(counties_i, index = fips)

# input Data to a dataFrame
input_data_df = pd.DataFrame(input_col_vals, index = fips )
input_data_df['state'], input_data_df['county'] = states_S, counties_S # add state and county
input_data_df.white, input_data_df.black, input_data_df.asian, input_data_df.hispanic = input_data_df.white * 100, input_data_df.black * 100, input_data_df.asian * 100, input_data_df.hispanic * 100 # Make demogrpahics into percentages
input_data_df.where(input_data_df!= np.NaN, other=0) # Replace missing values with zero ( These ones were originally zero values )
impute_values(input_data_df, [ 'suicides', 'health', 'physically unhealthy (days)', 'mentally unhealthy (days)', 'low birthweight', 'excess drinking' ]) # Impute the missing values for these variables using the state mean

# Output Data to a dataFrame
output_data_df = pd.DataFrame(output_col_vals, index = fips )
output_data_df['state'], output_data_df['county'] = states_S, counties_S # add state and county
impute_values(output_data_df, [ 'homicides', 'unemployment', 'pol_shootings', 'pol_deaths', 'child poverty', 'poverty_rate', 'crime_rate']) # Impute the missing values for these variables using the state mean

# create combined dataframe for input and output variables
input_output_df = pd.concat([input_data_df, output_data_df], axis = 1)

# export input and output data to separate files
input_data_df.to_csv('output/input_data.csv')
output_data_df.to_csv('output/output_data.csv')
```

##### Cell 6/8: Unsuperivsed random forest and PCA

```python
### Identifying outliers using Unsupervised Random Forest (URF)
X = (outcome_data_df.drop(columns = ['state', 'county']) * 1000).replace(np.NaN, 0).astype(int) # making values larger (to be able to use them as integers) and replacing nans
X = np.nan_to_num(X)
a = urf(n_trees=500, max_depth=3).get_anomaly_score(X, knn=3) # Getting the anomaly scores
y_pred = np.ones(a.shape, dtype=int) # getting a list for the shape of the anomaly scores
outliers = np.argsort(a)[::-1][:int(0.30 * len(a))] # Sort by anomaly score and select 30$ most anomalous
y_pred[outliers] = 0 # select only non-outliers from the y_pred

# Add column with 'Outlier' identifier to outcome variable dataframe
outcome_data_df['Outlier'] = list(y_pred)

# Double mean center data + Put mean centered data into distance(/proximity) matrix
X_d_matrix = urf(n_trees=500, max_depth=3).get_distance(X) # calculate the proximity matrix
X_dc_d_matrix = pd.DataFrame( double_mean_centering(X_d_matrix) ) # perform double mean centering

# Perform PCA without outliers
Xpca = X_dc_d_matrix # Define X for pca
X_pca_obj = PCA(n_components=6) # determine number of components and initiliase fitter
X_pca_obj.fit(Xpca) # fit the data from Xpca
X_pca = PCA(n_components=6).fit_transform(Xpca) # Transform the data
pc1_str = ''.join(['PC1(', str(round(X_pca_obj.explained_variance_[0],2)), '%)']) # determine what PC1 will be called in the figure
pc2_str = ''.join(['PC2(', str(round(X_pca_obj.explained_variance_[1],2)), '%)']) # idem for PC2
pca_outcome_data = pd.DataFrame(data = X_pca, columns = [pc1_str, pc2_str, 'Z', 'A', 'B', 'C']) # label the collumns (artbitrarily for non-visulised)

# viz. pca plot
sns.set_theme(style="darkgrid") # plot theme
g = sns.jointplot(x=pc1_str, y=pc2_str,  data=pca_outcome_data, # select which data to visualise
                  kind="scatter", # type of plot
                  color="b", height=7) # color and dimension
g.savefig('output/pca.png') # export plot to png file
```
